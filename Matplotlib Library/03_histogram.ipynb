Histogram to Understand Feature Distribution


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
feature = np.random.normal(loc=50, scale=10, size=1000)

plt.hist(feature, bins=30, color='skyblue', edgecolor='black')
plt.title('Feature Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
